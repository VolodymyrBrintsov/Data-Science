```python
import scipy.stats as t
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
```

# Using Numpy to generate test

### Create Random Normal Distribution

```python
dist = np.random.normal(0, 1, 10000)
sns.kdeplot(dist, shade=True)
```

### Choose random variable(we achieve it's z-score)

```python
random_z_score = np.random.choice(dist)
random_z_score
```

### Using `scipy` calculate the probability of its appearence in the distribution

```python
z_score_probability = t.norm.cdf(random_z_score)
z_score_probability
```

### Create Kde Plot to show shaded area of probability

```python
kde = t.gaussian_kde(dist)
# plot complete kde curve as line
pos = np.linspace(dist.min(), dist.max(), 101)
plt.plot(pos, kde(pos))
# plot shaded kde only right of x=0.5
shade = np.linspace(dist.min(), z_score_probability, 101)
plt.fill_between(shade,kde(shade), alpha=0.5)

plt.ylim(0,None)
plt.show()
```

# Using Real-World data

```python
df = pd.read_csv("../data/genetherapy.csv")
df.head()
```

### Show expression of gene using Kde Plot

```python
dist = np.array(df['expr'])
sns.kdeplot(dist, shade=True)
```

### Using `sklearn.preprocessing.StandartScaler` transform data to `z-score`

```python
scaler = StandardScaler()
dist = scaler.fit_transform(dist.reshape(-1, 1))
dist
```

```python
plt.hist(dist, bins=25)
```

### Choose random variable and get it's z-score, then calculate probability of it

```python
z_score_indice = np.random.choice(len(dist), size=1)
z_score = dist[z_score_indice]
print(z_score_indice, z_score)
probability = t.norm.cdf(z_score)
probability
```
